```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('../Data/follower_counts.csv')
df_top1000 = df.sort_values('follower_count', ascending=False).head(1000)
print(df_top1000)

# Apply logarithmic scaling and normalization to the follower count column
log_follower_count = np.log10(df_top1000['follower_count'])
norm_follower_count = (log_follower_count - log_follower_count.min()) / (log_follower_count.max() - log_follower_count.min())

# Create a scatter plot of normalized follower count vs. writer rank, with color gradient based on follower count
plt.scatter(np.arange(len(norm_follower_count)), norm_follower_count, c=df_top1000['follower_count'], cmap='viridis')
plt.colorbar(label='Follower Count')
plt.xlabel('Rank')
plt.ylabel('Normalized Follower Count')
plt.title('Distribution of Follower Count for Top 1000 Writers')
plt.show()
```
